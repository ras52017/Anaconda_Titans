# Read in and Standardize Diabetes Data

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
diabetes_binary_df = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2021_withBMI.csv")
diabetes_binary_df
```

```python
# Drop non-useful _STATE column
diabetes_columnDrop_df = diabetes_binary_df.drop(columns=['_STATE'], axis=1)
diabetes_columnDrop_df.head()
```

```python
# Check for unique values in each colum
diabetes_columnDrop_df.nunique()
```

```python
diabetes_columnDrop_df.dtypes
```

```python
# Convert GenHlth datatype for OneHotEncoding
diabetes_columnDrop_df['GenHlth'] = diabetes_columnDrop_df['GenHlth'].astype(str)
```

```python
# Convert MentHlth datatype for OneHotEncoding
diabetes_columnDrop_df['MentHlth'] = diabetes_columnDrop_df['MentHlth'].astype(str)
```

```python
# Convert MentHlth datatype for OneHotEncoding

diabetes_columnDrop_df['RACE'] = diabetes_columnDrop_df['RACE'].astype(str)

print(diabetes_columnDrop_df.dtypes)
```

```python
# Generate our categorical variable lists
categorical_lists = diabetes_columnDrop_df.dtypes[diabetes_columnDrop_df.dtypes == "object"].index.tolist()
categorical_lists
```

```python
# Generate our categorical variable lists
categorical_lists = diabetes_columnDrop_df.dtypes[diabetes_columnDrop_df.dtypes == "object"].index.tolist()
categorical_lists
```

```python
# Source for import warnings use: https://github.com/scikit-learn/scikit-learn/issues/21858
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(diabetes_columnDrop_df[categorical_lists]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(categorical_lists)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
merged_df = diabetes_columnDrop_df.copy()
merged_df = merged_df.merge(encode_df, left_index=True, right_index=True)
merged_df = merged_df.drop(categorical_lists, axis=1)
merged_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = merged_df.Diabetes_binary
X = merged_df.drop(["Diabetes_binary"], axis=1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=95, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

print(X_train_scaled.shape)
print(X_test_scaled.shape)
```

# Compile, Train, and Evaluate the Model

```python
# Define the logistic regression model
log_classifier = LogisticRegression(solver="lbfgs",max_iter=1000)

# Train the model
log_classifier.fit(X_train_scaled,y_train)

# Evaluate the model
y_pred = log_classifier.predict(X_test_scaled)
print(f" Logistic regression model accuracy: {accuracy_score(y_test,y_pred):.3f}")
```

```python
# Define the basic neural network model
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units=16, activation="relu", input_dim=65))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
